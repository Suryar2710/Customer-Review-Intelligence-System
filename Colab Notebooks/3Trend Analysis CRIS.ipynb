```python
df_exploded = pd.read_csv("/content/drive/MyDrive/CRIS/df_exploded.csv")
```

```python
df_exploded["review_time"] = pd.to_datetime(df_exploded["timestamp"], unit="ms", errors="coerce")
df_exploded["review_month"] = df_exploded["review_time"].dt.to_period("M").astype(str)
```

```python
monthly_aspects = df_exploded.groupby(["review_month", "aspects"]).size().reset_index(name="count")
monthly_aspects.head()
pivot_table = monthly_aspects.pivot(index="review_month", columns="aspects", values="count").fillna(0)
pivot_table.tail()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

for col in pivot_table.columns:
    plt.plot(pivot_table.index, pivot_table[col], label=col)

plt.legend()
plt.xticks(rotation=45)
plt.title("Aspect Trends Over Time")
plt.show()
```

```python
top_aspects = df_exploded["aspects"].value_counts().head(4).index
pivot_top = pivot_table[top_aspects]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

for col in pivot_top.columns:
    plt.plot(pivot_top.index, pivot_top[col], label=col)

plt.legend()
plt.xticks(rotation=45)
plt.title("Top 4 Aspect Trends Over Time")
plt.show()
```

```python
pivot_smooth = pivot_top.rolling(window=3).mean()
```

```python
plt.figure(figsize=(12,6))

for col in pivot_smooth.columns:
    plt.plot(pivot_smooth.index, pivot_smooth[col], label=col)

plt.legend()
plt.xticks(rotation=45)
plt.title("Smoothed Aspect Trends (3-month rolling)")
plt.show()
```

```python
negative_df = df_exploded[df_exploded["rating"] <= 2]

monthly_negative = negative_df.groupby(["review_month", "aspects"]).size().reset_index(name="count")

pivot_negative = monthly_negative.pivot(index="review_month", columns="aspects", values="count").fillna(0)

top_neg = negative_df["aspects"].value_counts().head(4).index

pivot_neg_top = pivot_negative[top_neg]
pivot_neg_smooth = pivot_neg_top.rolling(3).mean()
```

```python
plt.figure(figsize=(12,6))

for col in pivot_neg_smooth.columns:
    plt.plot(pivot_neg_smooth.index, pivot_neg_smooth[col], label=col)

plt.legend()
plt.xticks(rotation=45)
plt.title("Top Negative Issue Trends Over Time")
plt.show()
```
